## Extract all

```python
import os
from langchain_community.document_loaders import PyPDFLoader

# Folder containing your PDFs
pdf_folder = "/Users/halt/dev/school/nlp/OneDrive_1_2-11-2025"
output_folder = os.path.join(pdf_folder, "cleaned_txt")
os.makedirs(output_folder, exist_ok=True)

# Paragraphs to remove exactly
exact_removal = {
    "Fulbright University Vietnam Ground Floor, 105 Ton Dat Tien, Tan Phu, Quan 7, Ho Chi Minh City"
}

# Loop through each PDF
for filename in os.listdir(pdf_folder):
    if filename.lower().endswith(".pdf"):
        pdf_path = os.path.join(pdf_folder, filename)
        print(f"Processing {pdf_path}...")

        # Load PDF
        loader = PyPDFLoader(pdf_path)
        pages = loader.load()

        all_paragraphs = []

        for i, page in enumerate(pages):
            raw_paragraphs = [p.strip() for p in page.page_content.split("\n") if p.strip()]

            filtered = []
            for p in raw_paragraphs:
                if p in exact_removal or p.isdigit() or p.startswith("INTERNAL"):
                    continue
                filtered.append(p)

            # Merge lowercase first word with previous paragraph
            if i > 0 and filtered:
                first_word = filtered[0].split()[0]
                if not first_word[0].isupper():
                    all_paragraphs[-1] += " " + filtered[0]
                    filtered = filtered[1:]

            all_paragraphs.extend(filtered)

        # Save to .txt file
        output_path = os.path.join(output_folder, filename.replace(".pdf", ".txt"))
        with open(output_path, "w", encoding="utf-8") as f:
            for idx, para in enumerate(all_paragraphs, 1):
                f.write(f"Paragraph {idx}:\n{para}\n\n")

        print(f"Saved cleaned text to: {output_path}")
```

## Load Chroma

```python
from langchain_openai.embeddings import OpenAIEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
```

```python
import fitz  # PyMuPDF
import re
from sentence_transformers import SentenceTransformer
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.docstore.document import Document

# === Step 1: Clean and merge PDF pages ===
def load_and_clean_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    exact_removal = {
        "Fulbright University Vietnam Ground Floor, 105 Ton Dat Tien, Tan Phu, Quan 7, Ho Chi Minh City"
    }

    all_paragraphs = []

    for i, page in enumerate(doc):
        raw_paragraphs = [p.strip() for p in page.get_text().split("\n") if p.strip()]

        # Step 2: Filter
        filtered = []
        for p in raw_paragraphs:
            if p in exact_removal:
                continue
            if p.isdigit():
                continue
            if p.lower().startswith("internal"):
                continue
            if re.match(r"^\d+\s*\|\s*Page$", p):
                continue
            if re.match(r"^Page\s+\d+\s+of\s+\d+", p, re.IGNORECASE):
                continue
            filtered.append(p)

        # Step 3: Merge paragraph split across pages
        if i > 0 and filtered:
            first_word = filtered[0].split()[0] if filtered[0].split() else ""
            if first_word and not first_word[0].isupper():
                all_paragraphs[-1] += " " + filtered[0]
                filtered = filtered[1:]

        all_paragraphs.extend(filtered)

    return all_paragraphs

# === Step 2: Chunk the cleaned paragraphs ===
def chunk_paragraphs(paragraphs, chunk_size=500, chunk_overlap=100):
    splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    text = "\n\n".join(paragraphs)
    return splitter.split_text(text)

# === Step 3: Embed and store ===
def build_vectorstore(chunks, persist_path="./chroma_fulbright"):
    documents = [Document(page_content=chunk) for chunk in chunks]
    embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
    vectorstore = Chroma.from_documents(
        documents=documents,
        embedding=embedding_model,
        persist_directory=persist_path
    )
    vectorstore.persist()
    return vectorstore

# === Step 4: Main pipeline ===
def main():
    pdf_path = "/Users/halt/dev/school/nlp/OneDrive_1_2-11-2025/Academic-Policy_V5.1.pdf"  # replace with your path
    paragraphs = load_and_clean_pdf(pdf_path)
    chunks = chunk_paragraphs(paragraphs)
    vectorstore = build_vectorstore(chunks)
    print(f"✅ Stored {len(chunks)} chunks in Chroma at './chroma_fulbright'")

    # Example retrieval
    query = "What happens when a student is placed on academic probation?"
    docs = vectorstore.similarity_search(query, k=3)
    context = "\n\n".join([doc.page_content for doc in docs])
    print("\n=== Final Prompt to LLM ===\n")
    print(f"Context:\n{context}\n\nQuestion: {query}")

if __name__ == "__main__":
    main()
```
